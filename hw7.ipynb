```python
import pandas as pd
from citipy import citipy
import numpy as np
import requests as req
from weather2 import weather
```

```python
api_key = weather
```

```python
lats = np.random.uniform(low=-90.000, high=90.000, size=1)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1)
```

```python
lat_lng = (lats, lngs)
lat_lng
```

```python
city = citipy.nearest_city(lats,lngs)
city.city_name
```

```python
city.country_code
```

```python
target_url = 'http://samples.openweathermap.org/data/2.5/weather?lat=35&lon=139&appid=' + api_key
print(target_url)
```

```python
params = {
    'lat': lats,
    'lon': lngs,
    'units': 'metric',
    'APPID': api_key
}
```

```python
base_url = "http://api.openweathermap.org/data/2.5/forecast"
```

```python
response = req.get(base_url, params = params)
print(response.url)
```

```python
city_weather = response.json()
city_weather
```
